```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string

nltk.download('punkt')  # Download tokenization resources
nltk.download('stopwords')  # Download stopwords

cleaned_corpus
# Initialize the stemmer and stopwords
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

# Preprocessing function
def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text)
    
    # Remove punctuation and convert to lowercase
    tokens = [word.lower() for word in tokens if word.isalpha()]
    
    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
     
    # Join tokens back into a string
    preprocessed_text = ' '.join(tokens)
    
    return preprocessed_text

# Preprocess each tweet text
preprocessed_tweets = [preprocess_text(cleaned_corpus) for cleaned_corpus in cleaned_corpus]

# Display preprocessed tweet texts
for i, preprocessed_tweet in enumerate(preprocessed_tweets):
    print(f"Original Tweet: {cleaned_corpus[i]}")
    print(f"Preprocessed Tweet: {preprocessed_tweet}\n")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
# Assuming tweets is a list of your tweet texts
tweets =cleaned_corpus # List of tweet texts

# Step 1: Text Preprocessing (Tokenization and TF-IDF Vectorization)
vectorizer = TfidfVectorizer(max_features=1000)  # Limit the number of features for PCA
X_tfidf = vectorizer.fit_transform(tweets)
X_tfidf
# Step 2: Apply PCA on the TF-IDF data
pca = PCA(n_components=30)
X_pca = pca.fit_transform(X_tfidf.toarray())
X_pca.shape
```
